```python
from ultralytics import YOLO
import cv2

img = cv2.imread('./Dataset/D/17.jpg')

model = YOLO(model="best.pt")
result = model.predict(source=img, show=True, save=True, save_txt=True, conf=0.4)

for r in result:
    boxes = r.boxes
    for box in boxes:
        x1, y1, x2, y2 = box.xyxy[0] 
        label = r.names[int(box.cls)]
        confidence = box.conf
        print(f"Label: {label}, Confidence: {confidence}, Coordinates: {x1}, {y1}, {x2}, {y2}")
```
