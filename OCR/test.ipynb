```python
import cv2
import matplotlib.pyplot as plt
import easyocr
```

```python
img = cv2.imread('./1.jpg', 0)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
img_resized = cv2.resize(img, (0, 0), fx=2, fy=2, interpolation=cv2.INTER_CUBIC)
denoised = cv2.fastNlMeansDenoising(img_resized, None, 30, 7, 21)
plt.imshow(denoised)
```

```python
adaptive_thresholded = cv2.adaptiveThreshold(
    denoised, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2
)
plt.imshow(adaptive_thresholded)
```

```python
reader = easyocr.Reader(['fa'])
text = reader.readtext(img)
print(text[-1])
```
